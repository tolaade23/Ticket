# TICKET PATTERN IN PYTHON

In EITHER Java or Python, create a program that prints a ticket pattern after asking for input from the user about the number of tickets the user wants to print. The result of the program should use print statements to print out a pattern that matches the ticket template.




Note: the program should check that the value inputted by the user is a non-negative integer. If the number of tickets specified is greater than 2, then the tickets should be printed out one after another as shown in the below example (bold, underlined blue text is user input):

How many tickets should be printed?

2




 

```python
#Print out tickets pattern for specified non-negative integers.

#if number of ticket > 2, the tickets should be printed one after the other as indicated in the diagram in the question.

No_of_ticket = int(input("How many ticket should be printed? = ")) #Question to determine number of tickets to print. 
#Converted to integer as we need the value for the range in the loop condition.

print("  ") #space
 
print("\033[1;34;4m" + str(No_of_ticket ) + (" Ticket(s)") + "\033[0;0m") #Print number of tickets in bold, blue colour and underline.

print("  ") #Space

size = 14

for i in range(No_of_ticket): #The total number of repeated patterns for the specified number of ticket.
    
    for row in range (size): #There are 14 rows in the pattern.
        
        for col in range(size): #There are 14 columns in the pattern.
            
            if row%2==0 and (col==0 or col==13): 
                
                print("<", end = " ") #print symbol '<' if rows are even numbers i.e 0,2,6,.. and if on column 0 or 13.
                
            elif row%2!=0 and (col==0 or col==13): 
                
                print(">", end=" ") #print symbol '>' if rows are odd numbers i.e 1,3,5,.. and if on column 0 or 13.
                
            elif (col==1 or col ==12) or ((row==1 or row==12) and (col==2 or col <=11)):
                
                print("#", end = " ") #print symbol '#' if column=1 or 12  or when rows=1 or rows=12 and if column=2 or column=11
                
            elif row==0 and (col==2 or col<=11): 
                
                print("v", end=" ") #print symbol 'v' on rows 0 and when column = 2 or 11.
                
            elif row==13 and (col==2 or col<=11): 
                
                print("˄", end=" ") #print symbol '˄' on rows 13 and for column=2 or 11.
                
            elif(row==col): 
                
                print("#", end = " ") #print symbol'#' on rows that is on the same level as the column eg row 3, column 3
                
            elif (row==2 or row<=11) and (col==2 or col <=11): 
                
                print(" ", end = " ") #print blank space on rows=2 or=11 and when columns=2 or 11
                
        else: 
            
            print("") #process any other conditions that is not satisfied by the if and elif statement
```
